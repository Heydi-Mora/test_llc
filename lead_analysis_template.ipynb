# MORA Growth Analytics Lead Analysis

Use this notebook after exporting your Google Sheet leads as a CSV file. It helps you see which services people ask about and how inquiries change over time.

## 1. Export Your Google Sheet

In Google Sheets, go to `File > Download > Comma Separated Values (.csv)`. Save the file as `website_leads.csv` in the same folder as this notebook.

```python
from pathlib import Path
import pandas as pd

csv_path = Path("website_leads.csv")
leads = pd.read_csv(csv_path)
leads.head()
```

## 2. Clean Date Fields

```python
leads["Submitted At"] = pd.to_datetime(leads["Submitted At"], errors="coerce")
leads["Month"] = leads["Submitted At"].dt.to_period("M").astype(str)
leads.info()
```

## 3. Service Interest

```python
service_counts = leads["Service Interest"].fillna("Unknown").value_counts()
service_counts
```

```python
service_counts.plot(kind="bar", title="Lead Interest by Service", ylabel="Inquiries", xlabel="Service")
```

## 4. Monthly Inquiry Volume

```python
monthly_counts = leads.groupby("Month").size()
monthly_counts
```

```python
monthly_counts.plot(kind="line", marker="o", title="Website Inquiries by Month", ylabel="Inquiries", xlabel="Month")
```
